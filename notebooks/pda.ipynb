```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('https://s3-us-west-2.amazonaws.com/pcadsassessment/parking_citations.corrupted.csv')
```

```python
df.head()
```

```python
df.describe()
```

```python
df.count()
```

```python
# Meter Id, Marked Time, VIN will not be useful (too many NULL values). We already know Make is NULL half of the time
sns.heatmap(df.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# Copy data so we can drop rows
df_copy = df.copy()
```

```python
df_copy.drop(['Meter Id', 'Marked Time', 'VIN'], axis=1, inplace=True)
```

```python
df_copy.head()
```

```python
df_make_exists = df_copy[df_copy['Make'].notna()]
```

```python
df_make_exists.head()
```

```python
df_make_exists.count()
```

```python
df_no_make = df_copy[df_copy['Make'].isna()]
```

```python
df_no_make.head()
```

```python
df_no_make.count()
```

```python
sns.heatmap(df_make_exists.isnull(), yticklabels=False, cbar=False, cmap='viridis')
```

```python
# Let's look at our remaining missing values
df_make_exists.isnull().sum(axis = 0)
```

```python
df_make_exists.nunique()
```

```python
# Rather than impute values, for speed let's just drop all NAs from here
df_make_exists_dropna = df_make_exists.dropna()
```

```python
df_make_exists_dropna.isnull().sum(axis = 0)
```

```python
df_make_exists_dropna.nunique()
```

```python
df_make_exists_dropna.head()
```

```python
df_make_exists_dropna.count()
```

```python
# Split up data
X = df_make_exists_dropna.drop('Make', axis=1)
y = df_make_exists_dropna['Make']
```

```python
X.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
# Smaller data set to test functions
X_sm = pd.get_dummies(X.head(10000))
y_sm = y.head(10000)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_sm, y_sm, test_size=0.33, random_state = 42)
```

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
clf = RandomForestClassifier(n_estimators=100)
clf.fit(X_train, y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn import metrics
```

```python
print("Accuracy:", metrics.accuracy_score(y_test, y_pred))
```
