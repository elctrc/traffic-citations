```python
import pandas as pd
```

```python
df = pd.read_csv('../parking_citations.corrupted.csv')
```

```python
# Remove all NA values for Make
df = df[df['Make'].notna()]
```

```python
df.head()
```

```python
# calculate top 25 most common 'makes'
# This runs in less than one second but it's not the most intuitive piece of code
# common_makes = pd.DataFrame(df.groupby('Make')['Make'].count().sort_values(ascending=False).head(25))
common_makes = df.groupby('Make')['Ticket number'].count().sort_values(ascending=False).head(25)
```

```python
common_makes.index
```

```python
makes_list = [make for make in common_makes.index]
```

```python
makes_list
```

```python
# calculate most common 'Color' for each 'Make'
grouped = df.groupby(['Make', 'Color']).size().reset_index(name='counts')
```

```python
grouped.head(20)
```

```python
common_color = grouped.loc[grouped.reset_index().groupby(['Make'])['counts'].idxmax()]
```

```python
common_color.head(20)
```

```python
# find the first ticket issued for each 'Make'
df.groupby('Make')['Issue Date'].min().reset_index()
```
